```python
import numpy as np
```

**O que são arrays numpy?**

```python
# vamos usar uma função do numpy para ler nossos arquivos
km = np.loadtxt('carros-km.txt')
km
```

```python
# faremos a mesma coisa com o arquivo contendo os anos dos veículos
# veja que o argumento dtype nos diz qual o tipo de dado deverá ser convertido
anos = np.loadtxt('carros-anos.txt', dtype=int)
anos
```

```python
# agora vamos calcular quando kilometros por ano cada carro andou
# para fazer isso basta dividirmos a kilometragem pela idade do carro
km_media = km / (2021 - anos)
km_media
```

```python
# verificando o tipo de dados do array que criamos
type(km_media)
```

**Operações com variáveis**

```python
ano_atual = 2021
ano_fabricacao = 2003
km_total = 44410
km_media = km_total / (ano_atual - ano_fabricacao)
km_media
```

>Ou seja, este carro fez em média 2467 km por ano

```python
# se quisermos calcular qual o total de km desse carro no próximo ano basta somarmos o total à média
# usar o += é equivalente a somar a variável com o valor que vem após o símbolo de igual
km_total += km_media

km_total
```

**Declaração múltipla**

```python
# vamos declarar as mesmas variáveis acimas, mas agora de uma maneira diferente
ano_atual, ano_fabricacao, km_total = 2021, 2003, 44410
```
